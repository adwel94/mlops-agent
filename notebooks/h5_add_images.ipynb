# h5_add_images

액션 궤적(.h5)을 리플레이하면서 카메라 이미지를 입혀 데이터셋 .h5 를 만든다.

| 파라미터 | 설명 | 기본값 |
|---|---|---|
| `task` | 태스크 이름 | `"PickCube-v1"` |
| `count` | 리플레이할 에피소드 수 | `100` |
| `traj_path` | 입력 궤적 경로. 생략 시 `data/datasets/<task>/motionplanning.h5` (task_to_h5 출력) | `None` |
| `obs_mode` | 기록할 관측 모드 | `"rgb"` |

fetch 출력을 쓰려면 `traj_path` 에 `trajectory.h5` 를 지정. 반환: 데이터셋 .h5 경로.

```python
import sys
from pathlib import Path
ROOT = Path.cwd().parent
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))
```

```python
from scripts.h5_add_images import run

# fetch_sample_h5 로 받은 궤적을 입력으로 사용
traj = ROOT / 'data' / 'datasets' / 'PickCube-v1' / 'trajectory.h5'
ds = run(task='PickCube-v1', count=3, traj_path=str(traj))
print(ds, '| exists:', ds.exists())
```

CLI: `python scripts/h5_add_images.py --task PickCube-v1 --count 3 --traj-path data/datasets/PickCube-v1/trajectory.h5`
