# task_to_h5

모션 플래닝으로 액션 궤적을 생성해 `data/datasets/<task>/motionplanning.h5` 로 저장한다. (mplib이 Linux 전용이라 WSL에서 구동)

| 파라미터 | 설명 | 기본값 |
|---|---|---|
| `task` | 생성할 태스크. `SUPPORTED_TASKS` 중 하나 | `"PickCube-v1"` |
| `count` | 생성할 성공 에피소드 수 | `100` |
| `only_success` | 성공한 에피소드만 저장 | `True` |
| `obs_mode` | 보통 `"none"` (이미지 없는 원본 궤적) | `"none"` |
| `sim_backend` | 시뮬 백엔드 | `"cpu"` |

전제조건: WSL Ubuntu + conda `maniskill` env (SETUP.md). 반환: `motionplanning.h5` 경로.

```python
import sys
from pathlib import Path
ROOT = Path.cwd().parent
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))
```

```python
from scripts.task_to_h5 import run, SUPPORTED_TASKS
print(SUPPORTED_TASKS)

traj = run(task='ThreeColoredCubes-v1', count=3)
print(traj, '| exists:', traj.exists())
```
