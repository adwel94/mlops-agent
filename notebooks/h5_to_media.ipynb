# h5_to_media

데이터셋 .h5 에 기록된 RGB 프레임에서 영상/이미지를 추출한다. (시뮬레이터 안 돎)

**`video`** — 한 에피소드를 MP4로

| 파라미터 | 설명 | 기본값 |
|---|---|---|
| `traj_path` | 데이터셋 .h5 경로 | (필수) |
| `episode` | 에피소드 번호 | `0` |
| `fps` | 프레임레이트 | `30` |

**`preview`** — 여러 에피소드를 PNG 그리드로

| 파라미터 | 설명 | 기본값 |
|---|---|---|
| `traj_path` | 데이터셋 .h5 경로 | (필수) |
| `n` | 에피소드 수 | `9` |
| `frame` | `"first"`/`"mid"`/`"last"` 중 어느 프레임 | `"mid"` |

```python
import sys
from pathlib import Path
ROOT = Path.cwd().parent
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))
```

```python
from scripts.h5_to_media import video, preview
from IPython.display import Video, Image, display

ds = ROOT / 'data' / 'datasets' / 'PickCube-v1' / 'trajectory.rgb.pd_joint_pos.physx_cpu.h5'

mp4 = video(ds, episode=0)
display(Video(str(mp4), embed=True))
```

```python
png = preview(ds, n=9, frame='mid')
display(Image(str(png)))
```

CLI: `python scripts/h5_to_media.py video --traj-path <h5> --episode 0`
