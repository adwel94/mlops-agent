# check_maniskill_env

ManiSkill 환경이 "이 환경에서 하네스가 돌아가나?"를 검증하고, 고칠 수 있는 건 고치고(`--fix` → `requirements.txt` 설치), 못 하는 건(WSL·conda env 생성·시크릿) `SETUP.md` 를 가리켜 제안한다.

반환: `{"passed": bool, "checks": [{"name","level","detail","fix"}, ...]}`. `level` ∈ `pass|warn|fail`; `passed` 는 `fail` 이 하나도 없을 때 True (WSL/extras `warn` 은 전체 실패 아님).

```python
import sys
from pathlib import Path
ROOT = Path.cwd().parent
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))
```

```python
from scripts.check_maniskill_env import run

# 검증만. 복구까지: run(fix=True) (또는 CLI --fix) → requirements.txt 설치. WSL 프로브 생략: run(check_wsl=False).
result = run()
print('passed:', result['passed'])
for c in result['checks']:
    print(' ', c['level'].upper(), c['name'], '-', c['detail'])
    if c['level'] != 'pass' and c.get('fix'):
        print('      →', c['fix'])
```

CLI: `python scripts/check_maniskill_env.py [--fix] [--no-wsl]`
