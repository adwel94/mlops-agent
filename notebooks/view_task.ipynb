태스크 환경을 SAPIEN 데스크탑 뷰어 창으로 띄운다. 노트북 inline 아님 — **별도 OS 창**으로 뜨고, 로봇은 움직이지 않는다 (마우스로 카메라만 회전).

| 파라미터 | 설명 | 기본값 |
|---|---|---|
| `task` | 띄울 태스크 | `"PickCube-v1"` |
| `shader` | 렌더 셰이더 | `"default"` |
| `blocking` | `True`면 창 닫을 때까지 셀이 멈춤, `False`면 백그라운드로 띄우고 즉시 반환 | `False` |

`blocking=False`(기본)는 `Popen` 을 반환 → `p.terminate()` 로 닫는다.
