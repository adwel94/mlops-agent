# Setup Skill — Notebook Control

주어진 task의 ManiSkill 데모 데이터셋을 Hugging Face에서 다운로드.
**Idempotent**: 이미 받아져 있으면 스킵. `force=True` 시 재다운로드.

두 호출 패턴 모두 검증:
1. Function mode — `from scripts.setup import run`
2. Subprocess mode — 별도 프로세스 CLI 호출


## Setup — import path

```python
import sys
from pathlib import Path

PROJECT_ROOT = Path.cwd().parent
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))
print('project root:', PROJECT_ROOT)
```

## ① Function mode — 이미 받아진 task (스킵 확인)

```python
from scripts.setup import run, is_downloaded

task = 'PickCube-v1'
print('downloaded before call?', is_downloaded(task))
out = run(task=task)
print('returned path:', out)
print('exists:', out.is_dir())
```

## ② Subprocess mode — 동일 호출, 별도 프로세스

```python
import subprocess, sys as _sys

result = subprocess.run(
    [_sys.executable, str(PROJECT_ROOT / 'scripts' / 'setup.py'),
     '--task', 'PickCube-v1'],
    capture_output=True, text=True,
)
print('returncode:', result.returncode)
print(result.stdout.strip())
```

## ③ 다운로드된 task 디렉토리 구조 확인

```python
task_dir = Path.home() / '.maniskill' / 'demos' / 'PickCube-v1'
for sub in sorted(task_dir.iterdir()):
    if sub.is_dir():
        files = sorted(sub.iterdir())
        print(f'{sub.name}/')
        for f in files:
            size_kb = f.stat().st_size / 1024
            print(f'  {f.name:50s}  {size_kb:>10.1f} KB')
```
